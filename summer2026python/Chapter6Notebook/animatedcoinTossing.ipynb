```python
from matplotlib import animation
```

```python
def rollCoin(simulationCount = 1): # defaults to one if user leaves an argument blank
    heads: int = 1
    tails: int = 2

    rolls: list = [random.randrange(heads, tails + 1) for index in range(simulationCount)]
    values, frequencies = np.unique(rolls, return_counts = True)
    return values, frequencies, rolls
```

```python
simulationCount: int = 1000
```

```python
values, frequencies, rolls = rollCoin(simulationCount) # this argument decides the amount of flipping 
```
