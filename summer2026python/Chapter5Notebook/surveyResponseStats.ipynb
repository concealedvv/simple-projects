```python
# find the minimum, maximum, range, mean, median, mode, variance and standard deviation
```

```python
import numpy as np
```

```python
import statistics as stats
```

```python
surveyList: list = [1, 2, 5, 4, 3, 5, 2, 1, 3, 3, 1, 4, 3, 3, 3, 2, 3, 3, 2, 5]
```

```python
values, frequencies = np.unique(surveyList, return_counts = True)
```

```python
def displayFrequencies(valuesTable, frequenciesTable):
    for index, num in enumerate(values): # They have the same number elements so I can use the valuesTable index for frequenciesTable
        print(f" The value {num} has a frequency of {frequenciesTable[index]}")
```

```python
displayFrequencies(values, frequencies)
```

```python
def displayStatistics(surveyList): # shows minimum, maximum, range, mean, median, mode, variance and standard deviation
    statValues = [
        ("minimumValue", min(surveyList)),
        ("maximumValue", max(surveyList)),
        ("rangeValue", max(surveyList) - min(surveyList)),
        ("meanValue", stats.mean(surveyList)),
        ("medianValue", stats.median(surveyList)),
        ("modeValue", stats.mode(surveyList)),
        ("varianceValue", stats.variance(surveyList)),
        ("standardDeviationValue", stats.stdev(surveyList))
    ]

    for value in statValues:
        print(f" the {value[0]} has a value of {value[1]} ")


displayStatistics(surveyList)
```

```python
displayStatistics(surveyList)
```
